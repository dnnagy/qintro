- Alice has a q-state $\ket\psi = \alpha\ket 0 + \beta\ket 1$
- She wants to send her q-state to Bob
- Due to the no-cloning theorem, she can't make a copy of this state
- Alice can teleport her qubit to Bob using two classical bits and entanglement
